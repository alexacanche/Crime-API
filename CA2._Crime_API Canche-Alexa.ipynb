```python
import pandas as pd
import requests
import json
```

```python
def crime():
    """
    Function to that obtains a json dictionary 
    of all crime types in the data base

    Parameters:
    None
    
    Returns:
    A json dictionary
    """
    endpoint_url = 'https://hoyodecrimen.com/api/v1/crimes'
    res = requests.get(endpoint_url) #Do the request of the data and getting it 
    results = json.loads(res.content) #Changing the object into readable dictionary
    return results
```

```python
crime_types = crime()
crime_types
```

```python
df = pd.DataFrame(crime_types['rows']) #Create a data frame of the crime_types based on the rows
df
```

```python
def quadrants():
    """
    Function to that obtains a json dictionary 
    of all quadrants in the Database.

    Parameters:
    None
    
    Returns:
    A json dictionary
    """
    endpoint_url = ' https://hoyodecrimen.com/api/v1/cuadrantes'
    res = requests.get(endpoint_url) #Do the request of the data and getting it 
    results = json.loads(res.content) #Changing the object into readable dictionary
    return results
```

```python
quadrants_1 = quadrants()
quadrants_1
```

```python
df2 = pd.DataFrame(quadrants_1['rows']) #Create a data frame of the quadrants_1 based on the rows
df2
```

```python
def geographic_crimes(logitude, latitude):
    """
    Function to that obtains a json dictionary 
    of all the crimes within a certain distance 
    given the geographic coordinates of a point
    and the a distance of 2000 m.

    Parameters:
    None
    
    Returns:
    A json dictionary
    """
    endpoint_url = 'https://hoyodecrimen.com/api/v1/latlong/crimes/all/coords/{}/{}/distance/2000'.format(logitude, latitude)
    res = requests.get(endpoint_url) #Do the request of the data and getting it 
    results = json.loads(res.content) #Changing the object into readable dictionary
    return results
```

```python
zocalo_crime = geographic_crimes(-99.133334, 19.433735) #Locating all the crimes near the zocalo given the log/lat
zocalo_crime
```

```python
df3 = pd.DataFrame(zocalo_crime['rows']) #Create a table locating the crimes near 2 km to the Zocalo
df3
```

```python
def top_quadrants(start_date, end_date):
    """
    Function to that obtains a json dictionary 
    of the top ranked quadrants the highest crime 
    counts.

    Parameters:
    start_date: a str of the starting date of 
    the data.
    end_date: a str of the ending date of the 
    data.
    
    Returns:
    A list of top ranked quadrants the highest crime 
    counts.
    """
    parameters = {
        'start_date' : start_date, 
        'end_date' : end_date,
    }
    endpoint_url = 'https://hoyodecrimen.com/api/v1/cuadrantes/crimes/all/top/counts'
    res = requests.get(endpoint_url, params = parameters) #Do the request of the data and getting it 
    results = json.loads(res.content) #Changing the object into readable dictionary
    results = results['rows'] #Selecting from the dictionary the list for better use
    return results
```

```python
#Getting the data of the hole year of 2016
ranked_quadrants2016 = top_quadrants('2016-01', '2016-12') 
ranked_quadrants2016
```

```python
#Getting the data of the hole year of 2017
ranked_quadrants2017 = top_quadrants('2017-01', '2017-12') 
ranked_quadrants2017
```

```python
#Getting the data of the hole year of 2018
ranked_quadrants2018 = top_quadrants('2018-01', '2018-12')
ranked_quadrants2018
```

```python
#Concatenating the three lists into one to create the data frame
top_ranked_quadrants = ranked_quadrants2016 + ranked_quadrants2017 + ranked_quadrants2018
top_ranked_quadrants
```

```python
#Create a data frma with all the data of the list which contains the three years (2016,2017,2018)
table_top_quadrants = pd.DataFrame(top_ranked_quadrants) 

#Change the start_period column  to the name year and to str and only taking the year
table_top_quadrants['year'] = table_top_quadrants['start_period'].str.slice(start=0, stop=4)

#Selecting only the columns request by the amazing teacher Didier
table_top_quadrants = table_top_quadrants[['crime', 'cuadrante', 'population', 'rank', 'year']]

table_top_quadrants
```

```python
def sector_crimes(start_date, end_date):
    """
    Function to that obtains a json dictionary 
    of the count of crimes that occurred in all 
    sectors.

    Parameters:
    start_date: a str of the starting date of 
    the data.
    end_date: a str of the ending date of the 
    data.
    
    Returns:
    A list of count of crimes that occurred in all 
    sectors
    """
    parameters = {
        'start_date' : start_date, 
        'end_date' : end_date,
    }
    endpoint_url = 'https://hoyodecrimen.com/api/v1/sectores/all/crimes/all/period'
    res = requests.get(endpoint_url, params = parameters) #Do the request of the data and getting it 
    results = json.loads(res.content) #Changing the object into readable dictionary
    results = results['rows'] #Selecting from the dictionary the list for better use
    return results
```

```python
#Getting the data of the hole year of 2016
sector_crimes2016 = sector_crimes('2016-01', '2016-12') 
sector_crimes2016
```

```python
#Getting the data of the hole year of 2017
sector_crimes2017 = sector_crimes('2017-01', '2017-12') 
sector_crimes2017
```

```python
#Getting the data of the hole year of 2018
sector_crimes2018 = sector_crimes('2018-01', '2018-12') 
sector_crimes2018
```

```python
#Concatenating the three lists into one to create the data frame
total_sector_crimes = sector_crimes2016 + sector_crimes2017 + sector_crimes2018
total_sector_crimes
```

```python
#Create a data frma with all the data of the list which contains the three years (2016,2017,2018)
table_sector_crimes = pd.DataFrame(total_sector_crimes) 

#Change the start_period column  to the name year and to str and only taking the year
table_sector_crimes['year'] = table_sector_crimes['start_date'].str.slice(start=0, stop=4)

#Selecting only the columns request by the amazing teacher Didier
table_sector_crimes = table_sector_crimes[['count','crime', 'sector', 'year']]

table_sector_crimes
```
